````python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Load the data
data = np.load('power_trace.npy')[0:2000]
print(type(data))
print(f"shape: {data.shape}")

# Transform the data using STFT
# https://en.wikipedia.org/wiki/Short-time_Fourier_transform
f, t, Zxx = signal.stft(data, window='triang', nperseg=35)

fig, ax=plt.subplots(num=1, clear=True)
plt.pcolormesh(t, f, np.abs(Zxx))
plt.title('STFT Magnitude')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()

# Plot data
x = np.linspace(0, len(data), num=len(data))
fig, ax=plt.subplots(num=2, clear=True)
fig.suptitle('power_trace.npy')
ax.plot(x,data, linewidth=2.0)
plt.show()
````
